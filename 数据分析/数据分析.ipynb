```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('招聘信息 - 副本.csv',encoding='gbk')
df
```

```python
#查看工作地点最多的城市

df['工作地点'].value_counts()
```

```python
#查看招聘公司最多的类型

df['企业类型'].value_counts()
```

```python
df['公司类型'].value_counts()
```

```python
#查看哪个工作地点的平均工资最高

table = pd.pivot_table(df, index = ["工作地点"], values = ["平均薪资"])
table.sort_values("平均薪资", ascending = False)
```

```python
#查看学历与平均薪资之间的关系

table = pd.pivot_table(df, index = ["学历"], values = ["平均薪资"])
table.sort_values("平均薪资", ascending = False)
```

```python
#工作经验与平均薪资之间的关系
table = pd.pivot_table(df, index = ["工作经验"], values = ["平均薪资"])
table.sort_values("平均薪资", ascending = False)
```

```python
#不同城市，不同工作经验与薪资的关系

city_order = df.groupby('工作地点')['平均薪资'].mean().sort_values(ascending = False).head(30).index.tolist()
corr = df.pivot_table(index='工作地点',columns='工作经验',values='平均薪资')
corr = corr[['校招','无需经验','1年经验','2年经验','3-4年经验','5-7年经验','8-9年经验','10年以上经验']]
corr.loc[city_order]
```
