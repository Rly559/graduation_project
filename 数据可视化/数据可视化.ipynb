```python
import pandas as pd
from pyecharts.charts import *
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB

sns.set_style('white',{'font.sans-serif':['simhei','Arial']})
```

```python
df = pd.read_csv('招聘信息 - 副本.csv',encoding='gbk')
df
```

```python
#工作地点数量前十柱状图

fig,ax = plt.subplots(figsize=(12,8))

sns.countplot(x='工作地点',order=df['工作地点'].value_counts().head(10).index,data=df,color='r')

plt.box(False)

fig.text(x=0.04,y=0.9,s='各工作地点岗位数量',fontsize=32,weight='bold',color='k',backgroundcolor='#c5b783')

plt.tick_params(axis='both',which='major',labelsize=32)
ax.xaxis.grid(which='both',linewidth=0.5,color='#3c7f99')

plt.ylabel('岗位数量')
plt.xlabel('工作地点')
```

```python
#企业类型的数量分析

industry_index=df['企业类型'].value_counts().head(10).index
industry = df.loc[df['企业类型'].isin(industry_index),'企业类型']

fig,ax = plt.subplots(figsize=(12,8))

sns.countplot(y=industry.values,order=industry_index,color='y')

plt.box(False)

fig.text(x=0,y=0.9,s='企业类型的数量分析',fontsize=32,weight='bold',color='w',backgroundcolor='#c5b783')

plt.tick_params(axis='both',which='major',labelsize=32)
ax.xaxis.grid(which='both',linewidth=0.1,color='#3c7f99')
```

```python
fig,ax = plt.subplots(figsize=(12,8))

city_order = df.groupby('工作地点')['平均薪资'].mean().sort_values(ascending = False).head(10).index.tolist()

sns.barplot(x='工作地点',y='平均薪资',order=city_order,data=df,palette='RdBu_r')

plt.box(False)

fig.text(x=0.04,y=0.9,s='各城市的平均薪资对比',fontsize=32,weight='bold',color='w',backgroundcolor='#3c7f99')

plt.tick_params(axis='both',labelsize=16)
ax.yaxis.grid(which='both',linewidth=0.5,color='b')
ax.set_yticklabels(['','0.5k','1.0k','1.5k','2.0k','2.5k'])

plt.ylabel('平均薪资')
plt.xlabel('工作地点')
```

```python
fig,ax = plt.subplots(figsize=(12,8))

fig.text(x=0.04,y=0.9,s='一线城市的最高薪资对比',fontsize=32,weight='bold',color='w',backgroundcolor='#3c7f99')

sns.kdeplot(df[df['工作地点']=='北京']['最高薪资'],shade=True,label='北京')
sns.kdeplot(df[df['工作地点']=='上海']['最高薪资'],shade=True,label='上海')
sns.kdeplot(df[df['工作地点']=='广州']['最高薪资'],shade=True,label='广州')
sns.kdeplot(df[df['工作地点']=='深圳']['最高薪资'],shade=True,label='深圳')

plt.box(False)

plt.tick_params(axis='both',which='major',labelsize=16)
plt.xticks(np.arange(0,10,1),[str(i)+'w' for i in range(0,10,1)])
plt.yticks([])
plt.legend(fontsize = 'xx-large',fancybox=None)

plt.xlabel('')
plt.ylabel('')
```

```python
#不同城市，不同工作经验与薪资的关系

city_order = df.groupby('工作地点')['平均薪资'].mean().sort_values(ascending = False).head(10).index.tolist()
corr = df.pivot_table(index='工作地点',columns='工作经验',values='平均薪资')
corr = corr[['校招','无需经验','1年经验','2年经验','3-4年经验','5-7年经验','8-9年经验','10年以上经验']]
corr.loc[city_order]

fig.ax = plt.subplots(figsize = (30,20))

sns.heatmap(corr.loc[df['工作地点'].value_counts().head(10).index],cmap='RdBu_r',center=20,annot=True,annot_kws={'fontsize':14})

plt.tick_params(axis='x',which='major',labelsize=32)
plt.tick_params(axis='y',which='major',labelsize=32)

plt.xlabel('')
plt.ylabel('')
```

```python
#学历要求与岗位的关系

education = df['学历'].value_counts()

y = education.values
y = y/sum(y) #归一化，不进行的话系统会自动进行

plt.figure(figsize = (10,10))
plt.title("学历要求",fontsize = 15)
patches,l_text,p_text = plt.pie(y, labels = education.index, autopct = "%.1f %%", colors = "bygr", startangle = 90)

for i in p_text: #通过返回值设置饼图内部字体
    i.set_size(15)
    i.set_color('w')

for i in l_text: #通过返回值设置饼图外部字体
    i.set_size(15)
    i.set_color('r')
    
plt.legend() #图例
plt.show()
```

```python
from scipy.stats import norm
fig = plt.figure(figsize = (10,8))
ax1 = fig.add_subplot(111) #确认子图
n,bins,patches = ax1.hist(df["平均薪资"],bins = 100, color = 'm') #bins默认是10

ax1.set_ylabel("数量",fontsize = 15)
ax1.set_xlabel("万/月",fontsize = 15)
ax1.set_title("频率分布图",fontsize = 20)

#准备拟合
y = norm.pdf(bins,df["平均薪资"].mean(),df["平均薪资"].std()) #bins,mu,sigma
ax2 = ax1.twinx() #双轴
ax2.plot(bins,y,"b--")
ax2.set_ylabel("概率分布",fontsize = 15)
plt.show()
```

```python
#工作经验与岗位的关系

experience = df['工作经验'].value_counts()

y = experience.values
y = y/sum(y) #归一化，不进行的话系统会自动进行

plt.figure(figsize = (10,10))
plt.title("工作经验",fontsize = 15)
patches,l_text,p_text = plt.pie(y, labels = experience.index, autopct = "%.1f %%", colors = "bygr", startangle = 90)

for i in p_text: #通过返回值设置饼图内部字体
    i.set_size(15)
    i.set_color('w')

for i in l_text: #通过返回值设置饼图外部字体
    i.set_size(15)
    i.set_color('r')
    
plt.legend() #图例
plt.show()
```
