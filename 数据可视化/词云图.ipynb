```python
import jieba
import matplotlib.pyplot as plt
import wordcloud as wc
from PIL import Image
import numpy as np
```

```python
with open('福利词云图.txt',mode='r',encoding='utf-8') as fp:
    content = fp.read()
res = jieba.lcut(content)
text=''.join(res)

mask = np.array(Image.open('t014a4d7ffb3803587d.jpg'))

word_cloud = wc.WordCloud(font_path='msyh.ttc',mask = mask,background_color='white')
word_cloud.generate(text)
word_cloud.to_file('蜡笔小新.jpg')
plt.imshow(word_cloud)
plt.axis('off')
plt.show()
```
