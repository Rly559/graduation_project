```python
###薪资列处理

df.isnull().sum()
```

```python
###薪资缺失值处理

df.dropna(subset=['最高薪资'],inplace=True)
```

```python
df.isnull().sum()
```

```python
df['最低薪资']=df['薪资'].str.extract('^(.*?)-.*?万/月')
```

```python
df['最高薪资']=df['薪资'].str.extract('^.*?-(.*?)万/月')
```

```python
df['千/月最高薪资']=df['薪资'].str.extract('^(.*?)-.*?千/月')
```

```python
df['千/月最低薪资']=df['薪资'].str.extract('^.*?-(.*?)千/月')
```
